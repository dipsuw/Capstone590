```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv('../Data/2010 2015-ACS.csv', header=1)
```

```python
data.head()
```

```python
median_income =np.array(data['Median income (dollars); Estimate; Households'])
median_housevalue =np.array(data['Estimate; Median value (dollars)'])
bachelor_degree =np.array(data['Percent; Estimate; Percent bachelors degree or higher'])
```

```python
income_40th_perc = np.percentile(median_income, 40)
houseval_40th_perc = np.percentile(median_housevalue, 40)
bachelors_degree_top3rd_perc = np.percentile(bachelor_degree, 66)
```

```python
def add_col(row):
    return row['Estimate; Median value (dollars)']*1.08
```

```python
inflation_adj_median_house_value = data.apply(add_col, axis = 1)
```

```python
data['inflation_adj_median_house_value'] = data.apply(add_col, axis = 1)
```

```python
def label_def1(row):
   if row['Median income (dollars); Estimate; Households'] < income_40th_perc and row['Estimate; Median value (dollars)'] < houseval_40th_perc:
      return 'EligibleForGentrification'
   if row['Percent; Estimate; Percent bachelors degree or higher'] > bachelors_degree_top3rd_perc:
      return 'Gentrified'
   else:
      return
```

```python
#data.apply(label_def1, axis=1)
```

```python
data['def1'] = data.apply(label_def1, axis=1)
```

```python
median_income_40percent = np.percentile(median_income, 50)*0.4
```

```python
rent_increase = data['Renter-occupied housing units; Estimate; MONTHLY HOUSING COSTS AS A PERCENTAGE OF HOUSEHOLD INCOME IN THE PAST 12 MONTHS - $75,000 or more - 30 percent or more']
```

```python
median_rent_increase = np.percentile(rent_increase, 50)
```

```python
def label_def2(row):
   if row['Median income (dollars); Estimate; Households'] < median_income_40percent and row['Renter-occupied housing units; Estimate; MONTHLY HOUSING COSTS AS A PERCENTAGE OF HOUSEHOLD INCOME IN THE PAST 12 MONTHS - $75,000 or more - 30 percent or more'] > median_rent_increase:
      return 'CurrentlyGentrifying'
   else:
      return 'Gentrified'
```

```python
data['def2'] = data.apply(label_def2, axis=1)
```
